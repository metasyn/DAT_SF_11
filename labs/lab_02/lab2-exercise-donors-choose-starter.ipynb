```python
# Let's compare the completed vs expired for all the states
funding_count_by_state = projects.groupby([projects.school_state, projects.funding_status])['funding_status'].count()
funding_count_by_state
```

```python
# Count the number of projects posted by Year.
# Hint, create a new column for "date_posted" year using the mapping function - projects.date_posted.map(lambda dt: dt.year)
# Then do a distribution for projects posted by this new year column
projects['date_year'] = projects.date_posted.map(lambda dt: dt.year)
projects.date_year.hist()
```

```python
# Is the rest of the country catching up in posting these projects?
```

```python
# How long does it take from funding to completion - Has it changed over the years?
complete  = projects[projects.funding_status=='completed']
complete.tdelta = complete.date_completed - complete.date_posted
complete.tdelta.min() # there shouldn't be negative time
exclude = complete[complete.tdelta>'-1 days']
exclude['tdelta'] = complete.date_completed - complete.date_posted
exclude.tdelta.describe()
```

```python
from __future__ import division
exclude['date_delta'] = (exclude['date_completed'] - exclude['date_posted']) / np.timedelta64(1,'D')
```

```python
ax = exclude.groupby('date_year')['date_delta'].mean().plot()
ax.set_ylabel('Avg Days to Completion')
```
