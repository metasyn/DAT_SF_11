#3
Try PCA and see how much can you reduce the variable space.

How many Components did you need to explain 99% of variance in this dataset?

Plot the PCA variables to see if it brings out the clusters.

```python
from sklearn.decomposition import PCA
```

```python
dfpca = pd.DataFrame(columns=['components', 'explanations', 'model'])
```

```python
for i in range(0,14):
    p = PCA(n_components=i)
    p = p.fit(wine)
    pp = p.fit_transform(wine)
    dfpca.loc[i]= [p.components_, p.explained_variance_ratio_, pp]
```

```python
dfpca=dfpca[1:]
dfpca
```

```python
# Looks like you only need one
```

```python
for i in dfpca.model:
    pd.DataFrame(i).plot()
```

```python
for i in dfpca.model:
    pd.DataFrame(i).plot(kind='bar')
```

```python
for i in dfpca.model:
    pd.DataFrame(i).plot(kind='density')
```
